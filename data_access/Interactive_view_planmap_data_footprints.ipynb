Here we show how to list planmap's layers and visualize their footprints in an interactive view with folium

```python
from owslib import wfs
PLANMAP = 'https://geoserver.planmap.eu'
PLANMAP_WMS = PLANMAP + '/wms'
```

```python
from owslib.wms import WebMapService
wms = WebMapService(PLANMAP_WMS)
```

```python
target="mars"
basemap = "mars:Mars_MGS_MOLA_DEM_mosaic_global_463m"

m = folium.Map(location=[0,0],zoom_start=3, tiles=None,height=600, width=1200)

folium.raster_layers.WmsTileLayer(
    url=PLANMAP_WMS,
    name='basemap',
    format='image/png',
    layers=basemap,
    overlay=True,
    control=True,transparent=True, show=True
).add_to(m)


feature_group = folium.FeatureGroup("Footprints")

for id, meta in wms.contents.items():
    if id.startswith(target):

        bb = wms[id].boundingBox[:4]
        crs = wms[id].boundingBox[-1]
        print(f"{id} BB: {bb} CRS: {crs}")
        
        
        foot = folium.vector_layers.Rectangle(np.array(bb).reshape(2,2), opacity=0.4, popup=id)
#         foot.add_to(m)
        
        
        
        feature_group.add_child(foot) 

m.add_child(feature_group)
        
m.fit_bounds(bb)       


folium.LayerControl().add_to(m)

m
```
