```python
import pandas, geopandas
import numpy as np
```

```python
table = pandas.read_excel("units_APollo.xlsx")
table["label"] = table.Unit_Code
table
```

```python
dataframe = geopandas.GeoDataFrame.from_file("output/Apl_Strat/Apl_Strat.shp")
```

```python
tomatch = np.unique(dataframe.label)
testmatch = list(table.label)

for m in tomatch:
    if m not in testmatch:
        print(f"---> Check {m}")
    else:
        print(f"{m} is ok")
```

```python
fulldata = pandas.merge(table, dataframe)
fulldata = geopandas.GeoDataFrame(fulldata)
```
